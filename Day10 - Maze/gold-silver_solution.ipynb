## **Import Ã  file and convert it into a array of arrays**

```python
file_name = 'input.txt'

labyrinth = []
f = open(file_name, 'r').read().split()
for line in f:
     labyrinth.append(list(map(int,line)))
     print(line)
```

```python
border_r = len(labyrinth[0])
border_b = len(labyrinth)

reached_top = []
trail_count = 0
trail_rate = 0


def find_trailhead () :
    global reached_top
    trailhead_number = 0
    i = 0
    for line in labyrinth :
        j = 0
        for item in line :
            if (item == 0) :
                trailhead_number += 1
                path = 0
                reached_top = []
                
                trail_finder(i,j, path+1)
            j += 1
        i += 1
    return trailhead_number

def trail_finder (y,x, next_step):
    #print("next_step ->",next_step)
    global trail_rate
    orientation = [
        "top",
        "right",
        "bottom",
        "left"
    ]
    neighbor = [
        y!=0, 
        x!=border_r-1, 
        y!=border_b-1, 
        x!=0
    ]
    neighbor_coordinates = [
        [y-1,x],
        [y,x+1],
        [y+1,x],
        [y,x-1]
    ]
    for i in range (0,4) :
        #print("Voisin",orientation[i])
        if neighbor[i] :
            #print("   ",labyrinth[lab_neighbor[i][0]][lab_neighbor[i][1]])
            x = neighbor_coordinates[i][1]
            y = neighbor_coordinates[i][0]
            if labyrinth[y][x] == next_step :
                #print("   Find",orientation[i],"path to",next_step)
                if (next_step == 9):
                    trail_rate += 1
                    if unreached (x,y) :
                        add_trail_top(x,y)
                else :
                    trail_finder(neighbor_coordinates[i][0],neighbor_coordinates[i][1],next_step+1)
    return

def unreached (x,y):
    return not([x,y] in reached_top)

def add_trail_top (x,y):
    global reached_top
    global trail_count
    reached_top.append([x,y])    
    trail_count += 1
    return True

print(find_trailhead(),"trailhead found")
print(trail_count,"trails possibles")
print("Trail rate",trail_rate)
```
