```python
file_name = 'input.txt'
shemas = []
locks = []
keys = []

file = open(file_name, 'r').read().split("\n\n")

for shema in file:
    shemas.append(list(shema.split('\n')))

def clear_shema (shema) :
    shema.pop(0)
    shema.pop()
    for layer in shema :
        layer = list(map(str,layer))
    return shema

def convert_to_digit (shema) :
    pins_size = []
    for i in range (5) :
        sharp_number = 0
        for j in range (5) :
            if shema[j][i] == '#':
                sharp_number += 1
        pins_size.append(sharp_number)
    return pins_size

for shema in shemas :
    if shema[0] == '#####' :
        # Add a lock
        shema = clear_shema(shema)
        locks.append(convert_to_digit(shema))
    else :
        # Add a key
        shema = clear_shema(shema)
        keys.append(convert_to_digit(shema))
 

print("List of possible locks",locks)
print("List of possible keys",keys)
```

```python
def try_to_open (key, lock) :
    # print("try_to_open")
    global is_overlaps
    is_overlaps = False
    i = 0
    while i < 5 and not(is_overlaps_pins(key[i], lock[i])):
        i += 1
    # print(is_overlaps)
    return is_overlaps

def is_overlaps_pins(key_pin, lock_pin):
    # print("is_overlaps_pins")
    # print("Test : "+str(key_pin)+" + "+str(lock_pin)+" <= 5")
    if (key_pin + lock_pin <= 5) :
        return False
    else :
        global is_overlaps
        is_overlaps = True
        return True

is_overlaps = False
pears_locks_keys = []

for lock in locks :
    for key in keys :
        if not(try_to_open (key,lock)) :
            pears_locks_keys.append([lock,key])

print(len(pears_locks_keys),"pears found :")
for pear in pears_locks_keys :
    print(pear)
```
